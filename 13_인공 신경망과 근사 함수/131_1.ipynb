# 인공 신경망과 근사 함수

## 2차 함수 근사해 보기

```python
import numpy as np
import time
import matplotlib.pyplot as plt
```

```python
NUM_SAMPLES = 1000

np.random.seed(int(time.time()))
```

```python
xs= np.random.uniform(-2, 0.5, NUM_SAMPLES)
np.random.shuffle(xs)
print(xs[:5])
```

```python
ys = 2*xs**2 + 3*xs + 5
print(ys[:5])
```

```python
plt.plot(xs, ys, 'b.')
plt.show()
```
