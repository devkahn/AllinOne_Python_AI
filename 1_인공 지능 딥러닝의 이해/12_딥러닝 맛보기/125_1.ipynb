# 머신러닝 방식의 신경망 함수 생성과 사용

```python
import tensorflow as tf
import numpy as np
```

```python
xs = np.array([-1.0, 0.0, 1.0, 2.0, 3.0, 4.0])
ys = np.array([-2.0, 1.0, 4.0, 7.0, 10., 13.])
```

## Tensor 생성
### tf.keras.layers.InputLayer는 keras에서 제공하는 tensor를 생성하고, 입력 노드 개수를 정해준다.
#### Tensor는 3차워 이상의 행렬을 의미, 인공 신경망 구성 시 사용하는 자료형
### Dense 클래스를 이용하여 신경망 층을 생성
#### 여기서 Dense는 Y=X*w+b 식을 생성

```python
model = tf.keras.Sequential([
        tf.keras.layers.InputLayer(input_shape=(1,)),
        tf.keras.layers.Dense(1)
])
```

## 인공 신경망 생성
### model.compile을 호출하여 인공 신경망 구축
### 인공 신경망은 2개의 함수를 정해야함
#### loss : 손실 함수
##### mean_squared_error
#### optimizer : 최적화 함수
##### 확률적 하강함수 (sgd : stockastic gradient descent)

```python
model.compile(optimizer="sgd", loss='mean_squared_error')
model.fit(xs, ys, epochs=500)
```

```python
p = model.predict(x=np.array([10.0]))
print( "P :", p)
```
