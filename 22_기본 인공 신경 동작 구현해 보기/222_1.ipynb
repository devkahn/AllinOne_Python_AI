## 기본 인공 신경 동작 구현해보기

```python
x=2
t=10
w=3
b=1
```

```python
y=x*w + 1*b
print('y=%6.3f' %y)

yb = y-t
xb = yb*w
wb=yb*x
bb=yb*1
print('xb=%6.3f, wb=%6.3f, bb= %6.3f' %(xb,wb, bb))

lr = 0.01
w = w-lr*wb
b = b - lr*bb
print('x=%6.3f, w= %6.3f, b= %6.3f' %(x,w,b))
```

### 오차 값 계산하기

```python
x = 2
t = 10
w = 3
b = 1

for epoch in range(200) :
    print('epoch  = %d' %epoch)
    
    y = x*w + 1*b
    print('y = %6.3f' %y)
    
    E = (y-t)**2/2
    print('E=%.7f' %E)
    if E< 0.0000001:
        break
    
    yb = y - t
    xb = yb*w
    wb = yb*x
    bb = yb*1
    print('xb=%6.3f, wb = %6.3f, bb = %6.3f' %(xb, wb, bb))
    
    lr = 0.01
    w = w - lr*wb
    b = b - lr*bb
    print('x=%6.3f, w=%6.3f, b=%6.3f' %(x,w,b))
```
