### 2입력 3출력

```python
x1, x2 = .05, .10

t1, t2, t3 = .01, .99, .50

w1, w2 = .15, .20
w3, w4 = .25, .30
w5, w6 = .40, .55

b1, b2, b3 = .35, .45, .60

for epoch in range(2000):
    print('epoch = %d' %epoch)
    
    y1 = x1*w1 + x2*w2 + 1*b1
    y2 = x1*w3 + x2*w4 + 1*b2
    y3 = x1*w5 + x2*w6 + 1*b3
    print( 'y1, y2, y3 = %6.3f, %6.3f, %6.3f' %(y1, y2, y3))
    
    E = (y1 - t1)**2/2 + (y2 - t2)**2/2 + (y3 - t3)**2/2
    print ("E = %.7f" %E)
    if( E < 0.0000001):
        break
    
    y1b, y2b, y3b = y1 - t1, y2 - t2, y3 - t3
    x1b = y1b*w1 + y2b*w3 + y3b*w5
    x2b = y2b*w2 + y2b*w5 + y3b*w6
    w1b, w2b = x1*y1b, x2*y1b
    w3b, w4b = x1*y2b, x2*y2b
    w5b, w6b = x1*y3b, x2*y3b
    b1b, b2b, b3b= 1*y1b, 1*y2b, 1* y3b
    print( 'w1b, w3b, w5b = %6.3f, %6.3f, %6.3f' %(w1b, w3b, w5b))
    print( 'w2b, w4b, w6b = %6.3f, %6.3f, %6.3f' %(w2b, w4b, w6b))
    print( 'b1b, b2b, b3b = %6.3f, %6.3f, %6.3f' %(b1b, b2b, b3b))
    
    lr = 0.01
    w1, w3, w5 = w1 - lr*w1b, w3 - lr*w3b, w5 - lr*w5b
    w2, w4, w6 = w2 - lr*w2b, w4 - lr*w4b, w6 - lr*w6b
    b1, b2, b3 = b1 - lr*b1b, b2 - lr*b2b, b3 - lr*b3b
    print('w1, w3, w5 = %6.3f, %6.3f, %6.3f' %(w1,w3,w5))
    print('w2, w4, w6 = %6.3f, %6.3f, %6.3f' %(w2,w4,w6))
    print('b1, b2, b3 = %6.3f, %6.3f, %6.3f' %(b1,b2,b3))
    print()
```
