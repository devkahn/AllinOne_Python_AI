### 3입력 2출력

```python
x1, x2, x3 = .02, .05, .12
t1, t2 = .02, .98
w1, w2, w3 = .15, .20, .02
w4, w5, w6 = .27, .37, .52
b1, b2 = .12, .39


for epoch in range(2000):
    print(' epoch = %d' %epoch)
    
    y1 = x1*w1 + x2*w2 + x3*w3 + 1*b1
    y2 = x1*w4 + w2*w5 + w3*w6 + 1*b2
    print("  y1, y2 =%6.3f, %6.3f" %(y1, y2))
    
    E = (y1-t1)**2/2 + (y2-t2)**2/2
    print("  E = %.7f" %E)
    if E < 0.0000001:
        break
    
    y1b, y2b = y1 - t1, y2 - t2
    x1b, x2b, x3b = y1b*w1 + y2b*w4, y1b*w2 + y2b*w5, y1b*w3 + y2b*w6
    w1b, w2b, w3b = y1b*x1, y1b*w2, y1b*w3
    w4b, w5b, w6b = y2b*x1, y2b*w2, y2b*w3
    b1b, b2b = y1b*1, y2b*1
    print('    x1b, x2b, x3b = %6.3f, %6.3f, %6.3f' %(x1b, x2b, x3b))
    print('    w1b, w2b, w3b = %6.3f, %6.3f, %6.3f' %(w1b, w2b, w3b))
    print('    w4b, w5b, w6b = %6.3f, %6.3f, %6.3f' %(w4b, w5b, w6b))
    print('    b1b, b2b = %6.3f, %6.3f' %(b1b, b2b))
    
    lr = 0.01
    w1, w2, w3 = w1 - lr*w1b, w2 - lr*w2b, w3 - lr*w2b
    w4, w5, w6 = w4 - lr*w4b, w5 - lr*w5b, w6 - lr*w6b
    b1, b2 = b1 - lr*b1b, b2 - lr*b2b
    print('     w1, w2, w3 = %6.3f, %6.3f, %6.3f' %(w1,w2,w3))
    print('     w4, w5, w6 = %6.3f, %6.3f, %6.3f' %(w4,w5,w6))
    print('     b1, b2 = %6.3f, %6.3f' %(b1,b2))
    
    
    print()
```
