### 2입력 2은닉 3은닉 2출력

```python
i1, i2 = .01, .02
t1, t2 = .01, .99

w1, w3, w5 = .11, .30, .05
w2, w4, w6 = .12, .40, .06
b1, b2, b3 = .13, .41, .03

wh1, wh4 = .71, .01
wh2, wh5 = .81, .011
wh3, wh6 = .09, .021
bh1, bh2 = .47, .55

wo1, wo3 = .13, .51
wo2, wo4 = .14, .61
bo1, bo2 = .06, .71


for epoch in range(2000):
    print('epoch = %d' %epoch)
    
    h1 = i1*w1 + i2*w2 + 1*b1
    h2 = i1*w3 + i2*w4 + 1*b2
    h3 = i1*w5 + i2*w6 + 1*b3
    print("  h1, h2, h3 = %6.3f, %6.3f, %6.3f" %(h1, h2, h3))
    
    m1 = h1*wh1 + h2*wh2 + h3*wh3 + 1*bh1
    m2 = h1*wh4 + h2*wh5 + h3*wh6 + 1*bh2
    print("  m1, m2 = %6.3f, %6.3f" %(m1, m2))
    
    o1 = m1*wo1 + m2*wo2 + 1*bo1
    o2 = m1*wo3 + m2*wo4 + 1*bo2
    print("  o1, o2 = %6.3f, %6.3f" %(o1, o2))
    
    E = (o1 - t1)**2/2 + (o2 - t1)**2/2
    if E < 0.0000001:
        break
    
    o1b, o2b = o1 - t1, o2 - t2
    
    m1m, m2m = o1b*wo1 + o2b*wo3, o1b*wo2 + o2b*wo4
    wo1b, wo3b = o1b*m1m, o2b*m1m
    wo2b, wo4b = o1b*m2m, o2b*m2m
    bo1b, bo2b = o1b*1, o2b*1
    print("    wo1b, wo3b =%6.3f, %6.3f" %(wo1b, wo3b))
    print("    wo2b, wo4b =%6.3f, %6.3f" %(wo2b, wo4b))
    print("    bo1b, bo2b =%6.3f, %6.3f" %(bo1b, bo2b))
    
    h1b, h2b, h3b = m1m*wh1 + m2m*wh4, m1m*wh2 + m2m*wh5, m1m*wh3 + m2m*wh6
    wh1b, wh3b, wh5b = h1b*h1, h2b
    
    
    
    
    
    
    print()
```
