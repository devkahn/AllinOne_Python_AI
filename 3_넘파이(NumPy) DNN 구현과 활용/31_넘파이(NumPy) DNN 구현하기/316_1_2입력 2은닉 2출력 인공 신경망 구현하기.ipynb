# 넘파이(Numpy) DNN 구현과 활용
## 넘파이(NumPy) DNN 구현하기
### 2입력 2은닉 2출력 인공 신경망 구현하기

```python
import numpy as np

# numpy의 실수 출력 방법을 변경
# 소수점 이하 3자리까지 출력
np.set_printoptions(formatter={'float_kind':lambda x: "{0:6.3f}" .format(x)})
```

```python
X = np.array([[2]])
T = np.array([[27]])
W = np.array([[3]])
B = np.array([[1]])

for epoch in range(1000):
    print('\nepoch = %d' %epoch)
    
    Y = X @ W + B # @는 행렬의 곱
    print( 'Y = ', Y)
    
    E = np.sum((Y-T)**2/2)
    print('E = %.7f' %E)
    if E < 0.0000001:
        break
    
    Yb = Y - T
    Xb = Yb @ W.T
    Wb = X.T @ Yb
    Bb = 1 * Yb
    print(' Xb =\n', Xb)
    print(' Wb =\n', Wb)
    print(' Bb =\n', Bb)
    
    lr = 0.01
    W = W - lr * Wb
    B = B - lr * Bb
    print(' W =\n', W)
    print(' B =\n', B)
```
