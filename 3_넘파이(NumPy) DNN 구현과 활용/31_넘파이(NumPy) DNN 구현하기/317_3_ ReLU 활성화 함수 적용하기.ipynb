# 넘파이(Numpy) DNN 구현과 활용
## 넘파이(NumPy) DNN 구현하기
### 활성화 함수 적용하기
#### ReLU 함수 적용해 보기

```python
import numpy as np

# numpy의 실수 출력 방법을 변경
# 소수점 이하 3자리까지 출력
np.set_printoptions(formatter={'float_kind':lambda x: "{0:6.3f}" .format(x)})
```

```python
I = np.array([[.05, .10]])
T = np.array([[.01, .99]])

WH = np.array([[.15,.25], [.20,.30]])
BH = np.array([[.35,.35]])

WO = np.array([[.40,.50],[.45,.55]])
BO = np.array([[.60,.60]])

for epoch in range(10000):
    print(' epoch = %d' %epoch)
    
    H = I @ WH + BH
    H = (H>0)*H
    
    O = H @ WO + BO
    O = (O >0) *O
    print(' O = \n', O)
    
    E = np.sum((O-T)**2/2)
    print('  E = %.7f' %E)
    if E <  0.0000001:
        break
    
    Ob = O - T
    Ob  = Ob*(H>0)*1
    
    Hb = Ob @WO.T
    Hb = Hb*(H>0)*1
    
    WHb = I.T @ Hb
    BHb = 1*Hb
    WOb = H.T@Ob
    BOb = 1*Ob
    
    lr = 0.01
    WH = WH - lr * WHb
    BH = BH - lr * BHb
    WO = WO - lr * WOb
    BO = BO - lr * BOb
```
